```python
import requests as rq
from bs4 import BeautifulSoup
import csv
import pandas as pd
 
def partA() :
    header = ['job_title', 'company_name', 'salary', 'location', 'job_type', 'min_quals', 'desired_quals']
    data = []
 
    with open('JobSanaga.csv', 'w', newline='', encoding="UTF-8") as new:
        writer_obj = csv.writer(new)
        writer_obj.writerow(header)
 
        count = 0
        qualifications = []

        for i in range(1,2):
            url = 'https://www.snagajob.com/search?q=data+science&radius=50&page={}'.format(i)
            doc = rq.get(url)
            document = BeautifulSoup(doc.content, 'html.parser')
           
 
            for card in document.find_all('div', class_='job-overview'):
 
                count+=1
 
                company = card.find('h3', class_= 'body-md mb-0 mt-0 text-gray-700 ng-star-inserted').text
                job_title = card.find('span', class_= 'heading-300 text-gray-900 ng-star-inserted').text
 
                for element in card.find_all('div', class_='item-row grid grid-cols-[18px_1fr] gap-x-12 mt-8 ng-star-inserted'):
                    wage = card.find('div', attrs={"class": "item-content", "data-snagtag": "job-wage"})
                    if wage == None:
                        salary = 'N/A'
                    else:
                        salary = wage.find('span', class_= 'text-gray-900 ui-text-md ng-star-inserted').text
 
                    address = card.find('div', attrs={"class": "item-content", "data-snagtag": "job-address"})
                    if address == None:
                        location = 'N/A'
                    else:
                        location = address.find('div').text
                   
                    category = card.find('div', attrs={"class": "item-content", "data-snagtag": "job-categories"})
                    if category == None:
                        job_type = 'N/A'
                    else:
                        job_type = category.find('span', class_= 'text-gray-900 ui-text-md ng-star-inserted').text
 
                

                for link in card.find_all('a'):
                    content_url = link.get('href')
                    content_req = rq.get(content_url)
                    content_doc = BeautifulSoup(content_req.content, 'html.parser')
 
                    # for item in content_doc.find_all('div', class_= 'job-description relative prose body-md text-gray-900'):
                    #     quals = item.find('b')
 
                    #     if quals == None:
                    #         mins_quals = 'N/A'
                    #         desired_quals = 'N/A'
                    #     else:
                    #         subhead = quals.text.lower()
                    #         if 'qualifications' in subhead:
                    #             print('yay')

                    

                    if (content_doc.find_all(class_="job-description relative prose body-md text-gray-900")):
                        conttext = content_doc.find("jobs-description")
                        # if (len(conttext.split("Qualifications"))>1):
                        #     qualification = conttext[1]
                        # else:
                        #     qualification = ""
                        qual = conttext.find_all("div")
                        if(len(qual)>16):
                            qualifications.append(qual[18])
                        if(len(qual)==3):
                            if(qual[3].find_all("ul")):
                                qualifications.append(qual[3].find_all("ul"))
                            else:
                                qualifications.append("")
                        if(len(qual)==2):
                            if(qual[2].find_all("ul")):
                                qualifications.append(qual[2].find_all("ul"))
                            else:
                                qualifications.append("")
                        else:
                            qualifications.append("")
                    else:
                        qualifications.append("")

                    print(conttext)
                print(count, company)
                   
                
               
                job = [job_title, company, salary, location, job_type]
                writer_obj.writerow(job)
                # print(qualifications[count-1])

partA()
```
